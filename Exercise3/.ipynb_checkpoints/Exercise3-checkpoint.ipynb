```python
import librosa
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import get_window
from scipy.linalg import solve_toeplitz
from scipy.signal import freqz
from scipy.signal import lfilter
```

```python
def plot_signal(signal, sampling_freq, title):
    time = np.arange(signal.shape[0]) / sampling_freq
    plt.plot(time, signal)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title(title)
    plt.show()
```

#### 1. Load speech file

```python
s, sampling_rate = librosa.core.load("speech1.wav", sr=None)
```

#### 2. Select one unvoiced and one voiced speech segment and apply Hann window

```python
voiced = s[5200:5712]
plot_signal(voiced, sampling_rate, "Voiced segment")
```

```python
unvoiced = s[8200:8712]
plot_signal(unvoiced, sampling_rate, "Unvoiced segment")
```

```python
segment_length_ms = 32
segment_length = segment_length_ms * sampling_rate // 1000
v_analysis_window = get_window("hann", segment_length)
voiced *= v_analysis_window
unvoiced *= v_analysis_window

plot_signal(voiced, sampling_rate, "Voiced segment (with Hann window)")
plot_signal(unvoiced, sampling_rate, "Unvoiced segment (with Hann window)")
```

#### 3. Compute LP coefficients

```python
def compute_LP_coefficients(segment, M):
    autocorr_voiced = np.correlate(segment, segment, mode='same')
    
    # Calculate Phi_s
    autocorr_vector = autocorr_voiced[1:M+1]
    
    toeplitz_first_row = autocorr_voiced[0:M]
    LP_coefficients = solve_toeplitz((toeplitz_first_row, toeplitz_first_row), autocorr_vector)
    
    return LP_coefficients
```

```python
M = 12
a_voiced = compute_LP_coefficients(voiced, M)
print(a_voiced)
```

```python
a_unvoiced = compute_LP_coefficients(unvoiced, M)
print(a_unvoiced)
```

#### 4.a Make a plot of the frequency response of estimated vocal tract filter 

```python
def plot_freq_response(s, M, sampling_rate, plt_title):
    a = compute_LP_coefficients(s, M)
    w, h = freqz(1, np.concatenate(([1], a)), len(s), whole=True, fs=sampling_rate)
    
    fig, axs = plt.subplots(1, 2)
    fig.suptitle(plt_title)
    
    axs[0].plot(w, 10 * np.log10(np.abs(h)))
    axs[0].set_title("Magnitude response")
    axs[0].set_xlabel("Frequency (Hz)")
    axs[0].set_ylabel("Magnitude (dB)")
    
    axs[1].plot(w, np.angle(h))
    axs[1].set_title("Phase response")
    axs[1].set_xlabel("Frequency (Hz)")
    axs[1].set_ylabel("Phase (rad)")
    
    fig.tight_layout()
```

```python
plot_freq_response(voiced, M, sampling_rate, "voiced segment")
```

```python
plot_freq_response(unvoiced, M, sampling_rate, "unvoiced segment")
```

#### 5. Compute DFT of the windowed segments

```python
def plot_dft_and_filter(s, M, sampling_rate, plt_title):
    S = np.fft.rfft(s)
    freqs = np.fft.rfftfreq(len(s), 1 / sampling_rate)
    
    a = compute_LP_coefficients(s, M)
    w, h = freqz(1, np.concatenate(([1], a)), len(s), fs=sampling_rate)
    
    fig, axs = plt.subplots(1, 1)
    axs.plot(freqs, 10 * np.log10(np.abs(S)), label="DFT")
    axs.plot(w, 10 * np.log10(np.abs(h)), color="red", label="H(z)")
    axs.set_title("{title}, M = {M}".format(title = plt_title, M = M))
    axs.set_xlabel("Frequency (Hz)")
    axs.set_ylabel("Magnitude (dB)")
    axs.legend()
    
```

```python
plot_dft_and_filter(voiced, M, sampling_rate, "voiced segment")
```

```python
plot_dft_and_filter(unvoiced, M, sampling_rate, "unvoiced segment")
```

#### 6.a Compute residual signal

```python
def compute_residual(s, M, plt_title):
    a = compute_LP_coefficients(s, M)
    e = lfilter(np.concatenate(([1], a)), 1, s)
    
    fig, axs = plt.subplots(1, 1)
    axs.plot(e, color="red", label="Residual signal")
    axs.plot(s, label="Signal")
    axs.set_title(plt_title)
    axs.set_xlabel("Time (samples)")
    axs.set_ylabel("Amplitude")
    axs.legend()
```

```python
compute_residual(voiced, M, "voiced segment")
```

```python
compute_residual(unvoiced, M, "unvoiced segment")
```

M is the prediction filter polynomial order.


For voiced speech, there's usually one formant peak (local maxima) per kilohertz of bandwidth. You need two predictor terms to describe each formant peak. Using two extra LPC coefficients further minimizes residual energy.
Rule of thumb: number_of_LPC_terms = 2*bandwidth_in_KHz + 2

#### 9. pre-emphasis filter

```python
def pre_emphasis(s, alpha, M):
    pre_emphasized = np.copy(s)
    for i in range(s.shape[0]):
        pre_emphasized[i] = (pre_emphasized[i] - alpha * pre_emphasized[i-1])
    
    compute_residual(s, M, "voiced segment")
    compute_residual(pre_emphasized, M, "voiced segment - Pre-emphasized")
```

```python
pre_emphasis(voiced, 0.95, 12)
```

Without pre-emphasis, the residual signal oscillates within a larger (-0.1; 0.15) range, while the residual of a pre-emphasised signal stays closer (-0.075; 0.10) to the original signal.  

Pre-emphasizing the speech signal gives us a smaller error rate
